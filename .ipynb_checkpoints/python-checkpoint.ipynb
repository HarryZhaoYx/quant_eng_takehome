```python
def get_fin_data(ticker):
    ####################################################
    #### GET STOCK FUNDAMENTAL FINANCIAL DATA ####
    ####################################################
    endpoint = base_url + "financial_data/stock/search/?symbol={}".format(ticker.lower())
    stock_data = requests.get(endpoint, headers=headers).json()
    # json_formatted_str = json.dumps(stock_data, indent=2)
    fin_data = pd.DataFrame(stock_data["results"])
    return fin_data
```

```python

def get_mkt_data(ticker, start_date, end_date):
    ####################################################
    #### GET STOCK HISTORIC DATA ####
    ####################################################
    xignite_url = "https://globalhistorical.xignite.com/v3/xGlobalHistorical.json/GetGlobalHistoricalQuotesRange?"
    token = "4364FA47A6684516B601BE23F09DCC3C"
    params = {
        "_token": token,
        'Identifier': ticker,
        'StartDate': start_date,
        'EndDate': end_date,
        'IdentifierType': 'SYMBOL',
        'IdentifierAsOfDate': '10/10/2023',
        'AdjustmentMethod': 'All',
        'IncludeExtended': 'False',
    }
    
    response = requests.get(xignite_url, params=params)
    if response.status_code != 200:
        raise ValueError("Invalid response: %s" % response.text)
    return_data = json.loads(response.text)
    # print(return_data)

    # Extract the market data of this stock
    mkt_data = pd.DataFrame(return_data["HistoricalQuotes"])
    mkt_data["Ticker"] = return_data["Identifier"]
    col_order = ["Ticker"] + [col for col in mkt_data.columns if col != "Ticker"]
    mkt_data = mkt_data[col_order]
    return mkt_data
```

```python
def get_portfolio_data(portfolio, start_date = "2021-01-01", end_date = "2024-01-01"):
    # This is a function to combine all the data needed in a portfolio
    fin_data = pd.DataFrame()
    mkt_data = pd.DataFrame() 
    for stock in portfolio["stocks"]:
        _fin_data = get_fin_data(stock["ticker"])
        _mkt_data = get_mkt_data(stock["ticker"], start_date, end_date)

        fin_data = pd.concat([fin_data, _fin_data])
        mkt_data = pd.concat([mkt_data, _mkt_data])
    portfolio.update({"fundatmental": fin_data, "market": mkt_data})
    return portfolio
```

```python
def calculate_daily_performance(portfolio, start_date, end_date):
    portfolio = portfolio.copy()
    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)
    # Extract relevant information from the portfolio
    portfolio_name = portfolio['name']
    stocks_info = portfolio['stocks']
    mkt_data = portfolio['market']
    
    # Set the Date column as the datetime index
    mkt_data['Date'] = pd.to_datetime(mkt_data['Date'])
    mkt_data.set_index('Date', inplace=True)
    mkt_data = mkt_data.sort_index()
    
    # Truncate start data and end date of the data
    mkt_data = mkt_data.loc[start_date:end_date]

    # Calculate daily returns for each stock
    mkt_data['DailyReturn'] = mkt_data['PercentChange'] / 100

    # Initialize a DataFrame to store the weighted returns
    # weighted_returns = pd.DataFrame(index=mkt_data.index.unique())
    daily_returns = pd.DataFrame(index=mkt_data.index.unique())
    daily_weights = pd.DataFrame(index=mkt_data.index.unique())

    # Calculate the weighted returns for each stock based on sim-driven weights
    for stock_info in stocks_info:
        ticker = stock_info['ticker']
        daily_return = mkt_data[mkt_data["Ticker"] == ticker]['DailyReturn']
        # Calculate cumulative return for individual stock
        cumulative_return = np.empty_like(daily_return)
        cumulative_return[0] = daily_return.iloc[0]
        for i in range(1, len(daily_return)):
            cumulative_return[i] = (daily_return.iloc[i] + 1) * (1 + cumulative_return[i - 1]) - 1
        drifted_weight = stock_info['weight'] * (1 + cumulative_return)

        daily_returns[ticker] = daily_return
        daily_weights[ticker] = drifted_weight

    daily_weights = daily_weights.div(daily_weights.sum(axis=1), axis=0)
    
    weighted_returns = daily_weights * daily_returns

    # Calculate the portfolio's daily return by summing the weighted returns
    portfolio_return = pd.DataFrame()
    portfolio_return[portfolio_name] = weighted_returns.sum(axis=1)
    return {"returns" : portfolio_return, "weights" : daily_weights}

portfolio_1 = get_portfolio_data(portfolio_1)
portfolio_2 = get_portfolio_data(portfolio_2)

portfolio_1_results = calculate_daily_performance(portfolio_1, "2021-01-01", "2023-12-31")
portfolio_2_results = calculate_daily_performance(portfolio_2, "2021-01-01", "2023-12-31")
```

However, a critical aspect deserving attention is the maximum drawdown. Despite the tech stock portfolio's superior Sharpe ratio, it exposes investors to significant volatility, as evidenced by a substantial maximum drawdown of -46.06%. This substantial drawdown highlights a potential risk for investors seeking long-term growth with a desire to manage adverse risks. It underscores the importance of investors carefully considering risk tolerance and implementing risk mitigation strategies, even in portfolios exhibiting impressive returns.

```python
def annualized_return(returns):
    total_return = np.prod(1 + returns) - 1
    annualized_return = (1 + total_return) ** (252 / len(returns)) - 1
    return annualized_return

def annualized_volatility(returns):
    annualized_volatility = np.std(returns) * np.sqrt(252)
    return annualized_volatility

def sharpe_ratio(returns, risk_free_rate=0):
    excess_returns = returns - risk_free_rate
    sharpe_ratio = np.mean(excess_returns) / np.std(excess_returns)
    return sharpe_ratio

def max_drawdown(returns):
    cumulative_returns = np.cumprod(1 + returns) - 1
    peak = np.argmax(np.maximum.accumulate(cumulative_returns))
    trough = np.argmax(cumulative_returns[peak:])
    drawdown = (1+cumulative_returns.iloc[trough, 0])/(1+cumulative_returns.iloc[peak, 0])-1
    return drawdown 

def print_performance_table(portfolio_name, returns, risk_free_rate=0):
    annual_ret = annualized_return(returns)
    annual_vol = annualized_volatility(returns)
    sharpe = sharpe_ratio(returns, risk_free_rate)
    max_dd = max_drawdown(returns)

    # annual_ret = "{:.2f}%".format(annual_ret)
    # annual_vol = "{:.2f}%".format(annual_vol)
    # sharpe = "{:.2f}".format(sharpe)
    # max_dd = "{:.2f}%".format(max_dd)

    data = {
        "Annualized Return": annual_ret, 
        "Annualized Volatility": annual_vol,
        "Sharpe Ratio": sharpe, 
        "Max Drawdown": max_dd
    }

    data = pd.DataFrame(data)
    return data

performance_summary_1 = print_performance_table(portfolio_1["name"], portfolio_1_results["returns"])
performance_summary_2 = print_performance_table(portfolio_2["name"], portfolio_2_results["returns"])
performance_summary = pd.concat([performance_summary_1, performance_summary_2])

# Format the table
performance_summary = {key: [f"{value * 100:.2f}%" for value in values] for key, values in performance_summary.items()}
performance_summary = pd.DataFrame(performance_summary, index=[portfolio_1["name"], portfolio_2["name"]])
print(tabulate(performance_summary, headers="keys", tablefmt="pretty"))
```

### Scope Limitations and Additional Analyses:
Given the scope of this test, it's important to acknowledge the limitations. To refine and optimize the portfolio, additional analyses are recommended:

4. Weight Distribution Adjustment: Rebalance the portfolio weights based on the results of the factor attribution and optimization analyses.
5. Stress Testing: Assess the portfolio's performance under extreme market conditions to evaluate its resilience.
